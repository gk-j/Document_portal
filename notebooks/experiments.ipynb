#### 1. Data Ingestion

```python
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
import os
pdf_file_path=os.path.join(os.getcwd(),"data","sample.pdf")
```

```python
loader=PyPDFLoader(pdf_file_path)
```

```python
documents=loader.load()
len(documents)
```

#### This is the experimental way to determine the parameters there is no deterministic way 

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=150,length_function=len)
```

```python
docs=text_splitter.split_documents(documents)
len(docs)
```
